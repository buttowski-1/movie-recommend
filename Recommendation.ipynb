```python
import ast
```

```python
def convert(text):
  l=[]
  for i in ast.literal_eval(text):
    l.append(i['name'])
    # print(i)
    # break

  return l
```

```python
df['genres']=df['genres'].apply(convert)
```

```python
df['keywords']=df['keywords'].apply(convert)
```

```python
def convert_cast(text):
  l=[]
  count=0
  for i in ast.literal_eval(text):
    if count <3:
     l.append(i['name'])
    else:
      break
    # print(i)
    # break

  return l
```

```python

def convert_crew(text):
  l=[]
  for i in ast.literal_eval(text):
    if i['job']=='Director':
       l.append(i['name'])
    # print(i)
    # break

  return l
```

```python
df['cast']=df['cast'].apply(convert_cast)
```

```python
df['crew']=df['crew'].apply(convert_crew)
```

```python
def collapse(L):
  l=[]
  for i in L:
    l.append(i.replace(" ",""))
  return l
```

```python
df['cast'] = df['cast'].apply(collapse)
df['crew'] = df['crew'].apply(collapse)
df['genres'] = df['genres'].apply(collapse)
df['keywords'] = df['keywords'].apply(collapse)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv_vector = cv.fit_transform(df['tag']).toarray()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=5000, stop_words='english')
tfidf_vector = tfidf.fit_transform(df['tag']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cv_similarity = cosine_similarity(cv_vector)
tfidf_similarity = cosine_similarity(tfidf_vector)
```

```python
index=df[df['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie, similarity_matrix, top_n=5):
    index = df[df['title'] == movie].index[0]
    distances = sorted(
        list(enumerate(similarity_matrix[index])),
        reverse=True,
        key=lambda x: x[1]
    )
    return [df.iloc[i[0]].title for i in distances[1:top_n+1]]
```

```python
def avg_similarity_score(movie, similarity_matrix, k=5):
    index = df[df['title'] == movie].index[0]
    scores = sorted(similarity_matrix[index], reverse=True)[1:k+1]
    return sum(scores) / k
```

```python

# Option 2: If you wanted a label before the result
tfidf_res=recommend("Gandhi", tfidf_similarity)
print("TF-IDF:", tfidf_res)
```

```python

# Option 2: If you wanted a label before the result
cv_res=recommend("Gandhi", cv_similarity)
print("cv:",cv_res )
```

```python
common = set(cv_res) & set(tfidf_res)
print("Common recommendations:", common)
```

```python
print("TF-IDF Avg Similarity:",avg_similarity_score("Gandhi", tfidf_similarity))

print("Count Avg Similarity:",avg_similarity_score("Gandhi", cv_similarity))
```
