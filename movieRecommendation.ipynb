```python
import ast
```

```python
def convert(text):
  l=[]
  for i in ast.literal_eval(text):
    l.append(i['name'])
    # print(i)
    # break

  return l
```

```python
ast.literal_eval('[{"id": 28, "name": "Action"}]')
```

```python
df['genres'][0]
```

```python
df['genres']=df['genres'].apply(convert)
```

```python
df.head(1)
```

```python
df['keywords']=df['keywords'].apply(convert)
```

```python
def convert_cast(text):
  l=[]
  count=0
  for i in ast.literal_eval(text):
    if count <3:
     l.append(i['name'])
    else:
      break
    # print(i)
    # break

  return l
```

```python

def convert_crew(text):
  l=[]
  for i in ast.literal_eval(text):
    if i['job']=='Director':
       l.append(i['name'])
    # print(i)
    # break

  return l
```

```python
df['cast']=df['cast'].apply(convert_cast)
```

```python
df['crew']=df['crew'].apply(convert_crew)
```

```python
df.head(1)
```

```python
df.sample(5)
```

```python
def collapse(L):
  l=[]
  for i in L:
    l.append(i.replace(" ",""))
  return l
  
```

```python
df['cast'] = df['cast'].apply(collapse)
df['crew'] = df['crew'].apply(collapse)
df['genres'] = df['genres'].apply(collapse)
df['keywords'] = df['keywords'].apply(collapse)
```

```python
df['overview']=df['overview'].apply(lambda x: x.split(" "))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(df['tag']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
  index = df[df['title'] == movie].index[0]
  # distances=sorted
  distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
  for i in distances[1:6]:
        print(df.iloc[i[0]].title)
```
